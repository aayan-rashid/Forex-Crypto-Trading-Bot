### The next step, is to define the signal function.

Here, I am using the concept of Bullish and Bearish engulfings

```python
def signal_generator(df):
    open = df.Open.iloc[-1]
    close = df.Close.iloc[-1]
    previous_open = df.Open.iloc[-2]
    previous_close = df.Close.iloc[-2]

    # Bearish Englufing Pattern
    if (open>close and
    previous_open<previous_close and
    close<previous_open and
    open>=previous_close):
        return 1

    # Bullish Engulfing Pattern
    elif (open<close and
        previous_open>previous_close and
        close>previous_open and
        open<=previous_close):
        return 2

    # No clear pattern found !!
    else:
        return 0

```

#### Define function to collect the candles

```python

def get_candles(n):
    client = CandleClient(access_token, real=False)               # real = False --> Demo Account
    collector = client.get_collector(Pair.EUR_USD, Gran.M15)      # Pair is the EUR/USD and Granularity is of 15 minutes
    candles = collector.grab(n)                                   # n is the number of candles to be collected
    return candles

candles = get_candles(10)
for candle in candles:
    print(float(str(candle.bid.o))>1.08)                          # opening price for each of the 'n' candles. Type-cast the string into float to compare
```

#### Define the LIVE trading methodology

```python
def trading_job():
    candles = get_candles(3)
    dfstream = pd.DataFrame(columns=['Open','Close','High','Low'])

    # Get the Open, Close, High and Low prices LIVE from the market for 'n' candles
    i=0
    for candle in candles:
        dfstream.loc[i, ['Open']] = float(str(candle.bid.o))
        dfstream.loc[i, ['Close']] = float(str(candle.bid.c))
        dfstream.loc[i, ['High']] = float(str(candle.bid.h))
        dfstream.loc[i, ['Low']] = float(str(candle.bid.l))
        i=i+1

    # Convert the data into float type to make it comparable
    dfstream['Open'] = dfstream['Open'].astype(float)
    dfstream['Close'] = dfstream['Close'].astype(float)
    dfstream['High'] = dfstream['High'].astype(float)
    dfstream['Low'] = dfstream['Low'].astype(float)

    # Generate the signal for the last closed candles
    signal = signal_generator(dfstream.iloc[:-1,:])#


    # EXECUTING ORDERS
    client = API(access_token)


    # Take-Profit and Stop-Loss Ratio, i.e, Risk-Reward Ratio
    SLTPRatio = 2.

    # Calculate the previous closed candle price range
    previous_candleR = abs(dfstream['Open'].iloc[-2]-dfstream['Close'].iloc[-2])

    # Buy trades Take-Profts and Stop-Loss
    SLBuy = float(str(candle.bid.o))-previous_candleR
    TPBuy = float(str(candle.bid.o))+previous_candleR*SLTPRatio

    # Sell trades Take-Profits and Stop-Loss
    SLSell = float(str(candle.bid.o))+previous_candleR
    TPSell = float(str(candle.bid.o))-previous_candleR*SLTPRatio


    print(dfstream.iloc[:-1,:])
    print(TPBuy, "  ", SLBuy, "  ", TPSell, "  ", SLSell)

    signal = 2

    #Sell
    if signal == 1:
        mo = MarketOrderRequest(instrument="EUR_USD", units=-1000, takeProfitOnFill=TakeProfitDetails(price=TPSell).data, stopLossOnFill=StopLossDetails(price=SLSell).data)
        r = orders.OrderCreate(accountID, data=mo.data)
        rv = client.request(r)
        print(rv)
    #Buy
    elif signal == 2:
        mo = MarketOrderRequest(instrument="EUR_USD", units=1000, takeProfitOnFill=TakeProfitDetails(price=TPBuy).data, stopLossOnFill=StopLossDetails(price=SLBuy).data)
        r = orders.OrderCreate(accountID, data=mo.data)
        rv = client.request(r)
        print(rv)
```
